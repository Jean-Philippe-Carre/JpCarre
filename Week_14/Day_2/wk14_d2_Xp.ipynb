```typescript
/* Exercise 3 : The Temperature Converter

Instructions:

(1) Store a celsius temperature into a variable.

(2) Convert it to fahrenheit and console.log <temperature>°C is <temperature>°F.
Hint : Should you create another variable to hold the temperature in fahrenheit? (ie. point 2)
Hint: To convert a temperature from celsius to fahrenheit : Divide it by 5, then multiply it by 9, then add 32 */

let celsius = prompt("Enter a temperature in degrees: ");
console.log('You entered: ' + celsius);

let fahrenheit = (celsius * 9) / 5 + 32;
console.log(celsius + " deg" + " is " + fahrenheit + " fahrenheit.");
```
