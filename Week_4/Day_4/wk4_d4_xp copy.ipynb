```python

# EXERCISE 2:

# This project allows users to take a quiz to test their Star Wars knowledge.
# The number of correct/incorrect answers are tracked and the user receives different messages depending on how well they did on the quiz.

# Here is an array of dictionaries, containing those questions and answers:

# data = [
#     {
#         "question": "What is Baby Yoda's real name?",
#         "answer": "Grogu"
#     },
#     {
#         "question": "Where did Obi-Wan take Luke after his birth?",
#         "answer": "Tatooine"
#     },
#     {
#         "question": "What year did the first Star Wars movie come out?",
#         "answer": "1977"
#     },
#     {
#         "question": "Who built C-3PO?",
#         "answer": "Anakin Skywalker"
#     },
#     {
#         "question": "Anakin Skywalker grew up to be who?",
#         "answer": "Darth Vader"
#     },
#     {
#         "question": "What species is Chewbacca?",
#         "answer": "Wookiee"
#     }
# ]

# (1) Create a function that asks the questions to the user, and check his answers. Track the number of correct, incorrect answers. Create a list of wrong_answers
# (2) Create a function that informs the user of his number of correct/incorrect answers.
# (3) Bonus : display to the user the questions he answered wrong, his answer, and the correct answer.
# If he had more then 3 wrong answers, ask him to play again.

# **********************************************************************
# Part 1:
# **********************************************************************

data = [
    {
        "question": "What is Baby Yoda's real name?",
        "answer": "Grogu"
    },
    {
        "question": "Where did Obi-Wan take Luke after his birth?",
        "answer": "Tatooine"
    },
    {
        "question": "What year did the first Star Wars movie come out?",
        "answer": "1977"
    }, {
        "question": "Who built C-3PO?",
        "answer": "Anakin Skywalker"
    },
    {
        "question": "Anakin Skywalker grew up to be who?",
        "answer": "Darth Vader"
    },
    {
        "question": "What species is Chewbacca?",
        "answer": "Wookiee"
    }
]


def quiz():
    count_correct_answers = 0
    count_wrong_answers = 0
    wrong_answers = []

    for dict in data:
        keys = list(dict.keys())
        first_key = keys[0]
        first_value = dict[first_key]
        second_key = keys[1]
        second_value = dict[second_key]

        user_input = input(f"\n{first_key.capitalize()} : {first_value} ")
        if user_input.title() == second_value:
            count_correct_answers += 1
            print("Correct answer")
        else:
            count_wrong_answers += 1
            wrong_answers.append({
                "question": first_value,
                "user_input": user_input,
                "correct_answer": second_value,
            })
            print(f"Nope, definitely not {user_input}")

    print(f"\nYou have {count_correct_answers} correct answer(s) and {
        count_wrong_answers} wrong answer(s)")

    if wrong_answers:
        print("\nThese are your wrong answers:\n")
        for wrong in wrong_answers:
            print(f'Question: {wrong["question"]}')
            print(f'Your answer: {wrong["user_input"]}')
            print(f'Correct answer: {wrong["correct_answer"]}\n')

        else:
            pass

    if count_wrong_answers > 3:
        restart = input(
            "You got more than 3 wrong answers. Would you like to try again? (y/n): ")
        if restart.strip().lower() == 'y':
            quiz()


quiz()
```
